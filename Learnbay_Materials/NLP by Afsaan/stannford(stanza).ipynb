```python
import spacy_stanza
```

```python
doc = spacy_stanza.load_pipeline('en')
```

```python
analytics  = doc('Barack Obama was born in Hawaii. He was elected president in 2008.')
```

```python
for i in analytics:
    print(i.text , i.lemma_ , i.pos_)
```

```python
from spacy import displacy
```

```python
displacy.render(analytics , style = 'ent')
```
