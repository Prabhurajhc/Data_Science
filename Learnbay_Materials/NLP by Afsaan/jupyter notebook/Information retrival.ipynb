```python
import numpy as np
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.neighbors import KDTree
import joblib
```

```python
# read the dataste

person = pd.read_csv('famous_people.csv')
```

```python
person.head(10)
```

```python
# convert to vector form

tfidf = TfidfVectorizer()
train_tfidf = tfidf.fit_transform(person.Text).toarray()
```

```python
# adding the vectore as a column

person['tfidf'] = list(train_tfidf)
```

```python
person
```

```python
kdtree = KDTree(train_tfidf)

userinput = input('put something:')

trasnform = tfidf.transform([userinput]).toarray()
```

```python
userinput = input('put something:')

trasnform = tfidf.transform([userinput]).toarray()

distance , idx = kdtree.query(trasnform.reshape(1,-1), k=5)

for i, value in list(enumerate(idx[0])):
    print(f"Name: {person['Name'][value]}")
    print(f"URI: {person['URI'][value]}")      
```

```python
# models we havr to save

vector = joblib.dump(tfidf , 'tfidf_transform.pkl')
kd_model = joblib.dump(kdtree , 'kd_tree.pkl')
```
