# Iteration of array

```python
import numpy as np
```

```python
arr1 = np.array([10,11,12,13,14,14])
for i in arr1:
    print(i)
```

```python
arr2 = np.arange(25, 55).reshape(5, 6)
for i in arr2:
    print(i)
```

```python
arr2 = np.arange(25, 55).reshape(5, 6)
for i in arr2:
    for j in i:
        print(j)
```

```python
for i in arr2.ravel():
    print(i)
```

```python
for i in np.nditer(arr2, order= 'C'):    # order = 'C' -> row major form
    print(i)
```

```python
for i in np.nditer(arr2, order= 'F'):    # order = 'F' -> col major form
    print(i)
```

# Concatenation of array

```python
a = np.array([["New York","Maryland"],["Albany","Annapolis"]])

b = np.array([["Ohio","Arizona"],["Columbus","Phoenix"]])

c = np.array([["Ohio","Arizona"]])

d = np.array([["New York"],["Annapolis"]])\

d
```

```python
np.concatenate((a,b))   # Default axis = 0 => Concatenates along the row

# To do the concatenation along the row no of column should match
```

```python
np.concatenate((a,b), axis = 1)
# To do the concatenation along the col no of row should match
```

```python
np.concatenate((b,c))
```

```python
np.concatenate((b,d), axis = 1)
```

# Image manipulation using numpy and array

```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
astronaut = data.astronaut()
```

```python
plt.imshow(astronaut)
```

```python
astronaut.shape
```

# Cropping iamge

```python
plt.imshow(astronaut[100:, 50:, ::])
```

```python
# Split the image along the row
# plt.imshow(astronaut)
a1, a2 = np.split(astronaut, 2)
```

```python
plt.imshow(a1)
```

```python
plt.imshow(a2)
```

```python
# Split the image along the col
b1, b2 = np.split(astronaut, 2, axis = 1)
```

```python
plt.imshow(b1)
```

```python
plt.imshow(b2)
```

```python
# Joining the images along the row
plt.imshow(np.concatenate((a1, a2)))
```

```python
plt.imshow(np.concatenate((b1, b2), axis = 1))
```
